```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
from sklearn.preprocessing import StandardScaler
standart_scaler = StandardScaler()
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
model_SVC = SVC()
from sklearn.metrics import accuracy_score, r2_score
from sklearn.neighbors import KNeighborsClassifier
modelknnClas = KNeighborsClassifier(n_neighbors=5)
from sklearn.model_selection import cross_val_score, KFold
```

# DATA SETNI CHAQIRIB OLAMIZ 

```python
df = pd.read_csv("heart.csv")
df.info()
```

```python
df.head()
```

```python
def cleaner(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtypr=="object":
                df[col].fillna(df[col].mode()[0], inplace=True)
            else:
                df[col].fillna(df[col].mean(), inplace=True)
    return df 
```

```python
cleaner(df)
```

```python
df.info()
```

```python
def encodlash(df):    
    for col in df.columns:
        if df[col].dtype=="object":
            if df[col].nunique()<=6:
                dummies = pd.get_dummies(df[col], prefix="col", dtype=int)
                df = pd.concat([df.drop(columns=col), dummies,], axis = 1)
            else:
                encoder = LabelEncoder()
                df[col] = encoder.fit_transform(df[col])

    return df 
```

```python
encodlash(df)
```

```python
df.info()
```

```python
def scalinglash(df):
    num_col = df.select_dtypes(include=["float64", "int64"]).columns.drop("target")
    df[num_col]= standart_scaler.fit_transform(df[num_col])
    return df
```

```python
scalinglash(df)
```

```python
x = df.drop("target", axis=1)
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Algoritmimiz Support Vector Machine Classification 


```python
model_SVC
```

```python
model_SVC.fit(x_train, y_train)
```

```python
y_pred = model_SVC.predict(x_test)
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
score
```

```python
kn= KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model_SVC, x, y, cv=kn, scoring="accuracy")
print(scores)
```

```python
print(np.mean(scores). round(2))
```

```python
np.std(scores)
```
