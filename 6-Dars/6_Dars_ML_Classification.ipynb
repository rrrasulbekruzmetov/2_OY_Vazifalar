```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("tested.csv")
df.info()
```

```python
df.head()
```

```python
def cleaner(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="object":
                df[col].fillna(df[col].mode()[0], inplace=True)
            else:
                df[col].fillna(df[col].mean(), inplace=True)
    return df 
```

```python
df = cleaner(df)
```

```python
df.info()
```

```python
def encodlash(df):
    from sklearn.preprocessing import LabelEncoder
    encoder = LabelEncoder()    
    for col in df.columns:
        if df[col].dtype=="object":
            if df[col].nunique()<=6:
                dummies = pd.get_dummies(df[col], prefix="col", dtype=int)
                df = pd.concat([df.drop(columns=col), dummies,], axis = 1)
            else:
                encoder = LabelEncoder()
                df[col] = encoder.fit_transform(df[col])
    return df 
```

```python
df = encodlash(df)
```

```python
df.head()
```

```python
df.info() 
```

```python
def scalinglash(df):
    from sklearn.preprocessing import StandardScaler
    standart_scaler = StandardScaler()
    num_col = df.select_dtypes(include=["float64", "int64"])
    df[num_col]= standart_scaler.fit_transform(df[num_col])
    return df
```

```python
df = scalinglash(df)
```

```python
x = df.drop("Survived", axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Algoritm Decision Tree Classifier  


```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
```

```python
model_tree
```

```python
model_tree.fit(x_train, y_train)
```

```python
y_pred = model_tree.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.model_selection import cross_val_score, KFold
```

```python
score = accuracy_score(y_test, y_pred)
score
```

```python
kn= KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model_tree, x, y, cv=kn, scoring="accuracy")
print(scores)
```

```python
print(np.mean(scores). round(2))
```

```python
np.std(scores)
```

# Tuning 

# Hyperparameter

```python
dt_params={
    'max_depth':[2,4,6,8,10],
    'criterion':['gini','entropy'],
    'min_samples_split':[2,5,10]
}
```

```python
from sklearn.model_selection import GridSearchCV
dt = DecisionTreeClassifier(random_state=42)
grid_dt = GridSearchCV(estimator=dt, param_grid=dt_params, cv=5, scoring='accuracy')
grid_dt.fit(x_train, y_train)

best_dt = grid_dt.best_estimator_


y_pred_dt = best_dt.predict(x_test)
print("Decision Tree Best Parameters:", grid_dt.best_params_)
```

```python
score=accuracy_score(y_test,y_pred_dt)
score
```
