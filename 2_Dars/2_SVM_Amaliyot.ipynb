# Support Vector Machine (SVM)   Algoritmi bo`yicha amaliyot

```python
import pandas as pd 
df=pd.read_csv('adult.csv')
df.info()
```

```python
df.head()
```

# Data preprocessing 

```python
def cleaner(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype == "object":
                df[col].fillna(df[col].mode()[0], inplace=True)
            else:
                df[col].fillna(df[col].mean(), inplace=True)
    return df
```

```python
cleaner(df)
```

# Encoding 

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
def encodlash(df):
    for col in df.columns:
        if df[col].dtype=="object":
            if df[col].nunique()<=6:
                dummies = pd.get_dummies(df[col], prefix="col", dtype=int)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
            else:
                df[col]= encoder.fit_transform(df[col])
    return df
```

```python
df = encodlash(df)
```

```python
df.info()
```

# Scaling 

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler() 
```

```python
def scalinglash(df):
    num_col = df.select_dtypes(include=["int64"]).columns.drop("col_<=50K")
    df[num_col]= standard_scaler.fit_transform(df[num_col])
    return df 
```

```python
df = scalinglash(df)
```

```python
df.head()
```

# Data Setni qismlarga ajratamiz 

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("col_<=50K", axis=1)
y = df['col_<=50K']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR 
```

```python
model = SVR(kernel='rbf')
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

# Modelni Baholab Olamiz

```python
from sklearn.metrics import accuracy_score, r2_score
```

```python
class_score = r2_score(y_test, y_pred)
```

```python
class_score
```

```python
score = accuracy_score(y_test, y_pred)
```
